```python
# create a class
class Person:
    name = "Shyam"
    age = 5


ram = Person()
ram.name = "Ram"
print(ram.name)
print(ram.age)
```

A Class is like an object constructor, or a "blueprint" for creating objects.


# __init__()


```python
class Person:

    def __init__(self):
        self.name = 'Ram'
        self.age = 24

    def __init__(self, name, age):
        self.name = name
        self.age = age


ram = Person()
print(ram.name)

shyam = Person('Shyam', 24)
```
