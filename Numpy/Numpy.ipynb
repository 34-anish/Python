#Basics 


```python
# Initializing the numpy array
import numpy as np
a=np.array([1,2,3])
b=np.array([1.0,2.9,1])
c=np.array([['a','b','c'],['e','f','g']])
#.dtype returns the data type
print(a.dtype)
print(b.dtype)
print(c.dtype)

```

```python
# .itemsize returns number of bytes a data type occupies
print(a.itemsize)
print(b.itemsize)
print(c.itemsize)

```

```python
#.size returns the number of elements in the numpy array
print(a.size)
#.nbytes returns the total size occupied by a,b,c
print(a.nbytes)
print(b.nbytes)
print(c.nbytes)
```

Ways of repeating the elements

The axis of N-dimensional matrix can be defined with N+1 ways  
when the matrix is repeated (old matrix,number to repeat(n),axis =0) then the same matrix is repeated n times

```python
#With axis =0
a=np.array([[[1,2,3]]])
b=np.repeat(a,3,axis=0)
print(f"Shape of b is:{b.shape}")

b
```

```python
#With axis =1
b=np.repeat(a,3,axis=1)
print(f"Shape of b is:{b.shape}")
b
```

```python
#With axis =2
b=np.repeat(a,3,axis=2)
print(f"Shape of b is:{b.shape}")
b
```

```python
#Default axis
b=np.repeat(a,3)
print(f"Shape of b is:{b.shape}")
b
```

#Understanding axis

```python
ax1=np.arange(24).reshape(2,3,4)
sum=np.sum(ax1)
print(f"Sum without axis= {sum}")
print(f"ax1 === \n {ax1}")

#for axis =0 we do arr[0]+arr[1]
print(f"ax1[0] === \n {ax1[0]}")
print(f"ax1[1] === \n {ax1[1]}")
sum=np.sum(ax1,axis=0)
print(f"Sum when axis=0 is:\n {sum}")

#for axis = 1 we do arr[0][0]+arr[0][1]+arr[0][2]
print(f"ax1[0][0] === \n {ax1[0][0]}")
print(f"ax1[0][1] === \n {ax1[0][1]}")
print(f"ax1[0][2] === \n {ax1[0][2]}")
sum=np.sum(ax1,axis=1)
print(f"Sum when axis=1 is:\n {sum}")

#for axis = 2 we do arr[0][0][0]+arr[0][0][1]+arr[0][0][2]+arr[0][0][3]
print(f"ax1[0][0][0] === \n {ax1[0][0][0]}")
print(f"ax1[0][0][1] === \n {ax1[0][0][1]}")
print(f"ax1[0][0][2] === \n {ax1[0][0][2]}")
print(f"ax1[0][0][3] === \n {ax1[0][0][3]}")

sum=np.sum(ax1,axis=2)
print(f"Sum when axis=1 is:\n {sum}")
```
